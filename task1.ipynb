```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv('report.csv')
df
```

```python
df['Revenue Growth (%)'] = df.groupby('Company')['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby('Company')['Net Income'].pct_change() * 100
```

```python
df
```

```python
df['Assets Growth (%)'] = df.groupby('Company')['Total Assets'].pct_change() * 100
df['Liabilities Growth (%)'] = df.groupby('Company')['Total Liabilities'].pct_change() * 100
df['Cash Flow from Operations Growth(%)'] = df.groupby('Company')['Cash Flow from Operating Activities'].pct_change() * 100
```

```python
df.fillna(0, inplace=True)
df
```

```python
df.to_csv('final_data_report.csv')
```

```python
summary = df.groupby('Company').agg({'Total Revenue': 'sum', 'Net Income': 'sum', 'Total Assets': 'sum', 'Total Liabilities': 'sum', 'Cash Flow from Operating Activities': 'sum', 'Revenue Growth (%)': 'mean', 'Net Income Growth (%)': 'mean', 'Assets Growth (%)': 'mean', 'Liabilities Growth (%)': 'mean', 'Cash Flow from Operations Growth(%)': 'mean'}).reset_index()
```

```python
summary
```

```python
summary.to_csv('summary_report.csv')
```
